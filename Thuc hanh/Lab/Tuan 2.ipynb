```python
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
print(f'Tổng của {a} + {b} là: {a + b}')
print(f'Hiệu của {a} - {b} là: {a - b}')
print(f'Tích của {a} * {b} là: {a * b}')
print(f'Thương của {a} / {b} là: {a / b}')
print(f'Phần nguyên của {a} // {b} là: {a // b}')
print(f'Phần dư của {a} % {b} là: {a % b}')
```
