W4A7: Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n. 

```python
n = int(input("Nhập n: "))
i = 2
while i * i <= n:
    if n % i == 0:
        n //= i
    else:
        i += 1
print("Ước nguyên tố lớn nhất là:", n)
```

W4A8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input("Nhập n: "))
step = 0
while True: 
    s = str(n)
    if s == s[::-1]: 
        break
    n = n + int(s[::-1])
    step += 1
print(f"Số bước: {step}")
print(f"Giá trị palindrome: {n}")
```

W4A9: Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number). 

```python
n = int(input("Nhập n: "))
result = []
for x in range(1, int(n**0.5) + 1):
    sq = x * x
    s_sq = str(sq)
    if len(s_sq) == len(set(s_sq)):
        result.append(s_sq)
if result:
    print(" ".join(result))
else:
    print("(no number)")
```

W4A11: Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6. 

```python
n = int(input("Nhập n: "))
count = 0
for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1
        j = n // i
        if j != i and j % 2 == 0:
            count += 1
print(f"Số lượng ước số chẵn của {n} là: {count}")
```
