W3A1: Nhập vào số nguyên dương có đúng 3 chữ số, tìm và in ra số đảo ngược

```python
n = int(input())
if 100<=n<=999:
    dv = n%10
    chuc = (n//10)%10
    tram = n//100
    n_reserve = dv*100 + chuc*10 + tram
    print("Số đảo ngược:", n_reserve)
else:
    print("Không phải số có 3 chữ số")
```

W3A2: Nhập vào 1 số n, đại diện cho một tháng trong năm nhuận. In ra số ngày tương ứng.

```python
n = int(input())
match n:
    case 1|3|5|7|8|10|12:
        print(f"Tháng {n} có 31 ngày")
    case 4|6|9|11:
        print(f"Tháng {n} có 30 ngày")
    case 2:
        print(f"Tháng {n} có 29 ngày")
```

W3A3: Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input())
if n>0 and n & (n-1) == 0:
    print(f"{n} là lũy thừa của 2")
else:
    print(f"{n} không phải là lũy thừa của 2")
```
