W3A1: Nhập vào số nguyên dương có đúng 3 chữ số, tìm và in ra số đảo ngược

```python
n = int(input())
if 100<=n<=999:
    dv = n%10
    chuc = (n//10)%10
    tram = n//100
    n_reserve = dv*100 + chuc*10 + tram
    print("Số đảo ngược:", n_reserve)
else:
    print("Không phải số có 3 chữ số")
```

W3A3: Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input())
if n>0 and n & (n-1) == 0:
    print(f"{n} là lũy thừa của 2")
else:
    print(f"{n} không phải là lũy thừa của 2")
```

W3A10: Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".

```python
a, b, c = map(int,input().split())
if a + b > c and a + c > b and b + c > a:
    print("Tam giác hợp lệ")
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giác vuông")
    else: 
        print("Tam giác thường")
else:
    print("Không là ba cạnh của một tam giác")
```

W3A12: Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh

```python
n = int(input())
tien = 0
match n:
    case n if n<=50:
        tien = n*1500
    case n if n<=100:
        tien = 50*1500 + (n-50)*2000
    case n if n>100:
        tien = 50*1500 + 50*2000 + (n-100)*3000
print(f"Tiền điện phải trả cho {n} kWh là: {tien} đồng")
```

W3A13: Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng. 

```python
a, b = map(float,input().split())
if a != 0:
    x = -b / a
    print(f"Phương trình có một nghiệm duy nhất: x = {x:.2f}")
else:
    if b == 0:
        print("Phương trình có vô số nghiệm")
    else:
        print("Phương trình vô nghiệm")
```

W3A14: Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu

```python
N = float(input())
match N:
    case N if N >= 8.0:
        print("Học sinh xếp loại: Giỏi")
    case N if N >= 6.5:
        print("Học sinh xếp loại: Khá")
    case N if N >= 5.0:
        print("Học sinh xếp loại: Trung bình")
    case _:
        print("Học sinh xếp loại: Yếu")
```

W3A16: Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs d. 

```python
a, b, c, d = map(int, input().split())
if b//a==0:
    cong_boi = b//a
    if c == b*cong_boi and d == c*cong_boi:
        print("Đây là cấp số nhân theo đúng thứ tự đã cho")
else: 
    print("Đây không phải là cấp số nhân theo đúng thứ tự đã cho")
```
