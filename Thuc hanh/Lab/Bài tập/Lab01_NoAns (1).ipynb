### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def problem_4(n):
    count = 0
    for k in range(1, n):
        if k > (n - 1) / 2:
            count += 1
    return count

print(problem_4(9))
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
def problem_6(a, b, c):
    # Check if sides can form a valid triangle
    if a + b <= c or a + c <= b or b + c <= a:
        raise ValueError("The sides cannot form a triangle")
    
    # Heron's formula
    s = (a + b + c) / 2
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    return area

print(problem_6(3, 4, 5))
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def problem_8(x, n):
    if not isinstance(x, str) or len(x) != 1 or not x.isalpha():
        raise ValueError("Input must be a single alphabetic character")
    if x.islower():
        base = ord('a')
    else:
        base = ord('A')
    return chr(base + (ord(x) - base + n) % 26)

# some tests
print(problem_8('x', 3))  # wraps to 'a'
print(problem_8('Z', 2))  # 'B'
```
