Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    spaces = " " * (h - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
a = int(input("Nhập vào số nguyên dương a: "))
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i
if tong_uoc == a:
    print(f"{a} là số hoàn hảo.")
else:
    print(f"{a} không phải là số hoàn hảo.")
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) 
lớn nhất nhưng nhỏ hơn n 

```python
n = int(input("Nhập vào số nguyên n: "))
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k
print(f"Số k lớn nhất sao cho S(k) < {n} là: {k}")
print(f"Tổng S({k}) tương ứng là: {S}")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input("Nhập vào số nguyên dương A: "))
tong_s = 0
n = 0
while tong_s <= A:
    n += 1
    tong_s += 1 / n

# In kết quả
print(f"Số n nhỏ nhất thỏa mãn là: {n}")
print(f"Tổng S tương ứng là: {tong_s:.5f}")
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. 

```python
print("Nhập vào các số nguyên (Nhập -1 để dừng):")
numbers = []
while True:
    n = int(input("Nhập số: "))
    if n == -1:
        break
    numbers.append(n)
if len(numbers) > 0:
    max_val = max(numbers)
    min_val = min(numbers)
    print("--- Kết quả ---")
    print(f"Số lớn nhất trong dãy là: {max_val}")
    print(f"Số nhỏ nhất trong dãy là: {min_val}")
else:
    print("Bạn chưa nhập số nào hợp lệ.")
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ

```python
n = int(input("Nhập vào số nguyên dương n: "))
A = n
count_chan = 0
count_le = 0
tong_chu_so = 0
if n == 0:
    count_chan = 1
    tong_chu_so = 1
else:
    while n > 0:
        chu_so = n % 10
        if chu_so % 2 == 0:
            count_chan += 1
        else:
            count_le += 1
            
        tong_chu_so += 1
        n //= 10
print(f"--- Kết quả cho số {A} ---")
print(f"Số lượng chữ số chẵn: {count_chan}")
print(f"Số lượng chữ số lẻ: {count_le}")
print(f"Tổng số chữ số: {tong_chu_so}")
```

Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập vào số nguyên dương n: "))
temp = n
tong_chu_so = 0
while n > 0:
    chu_so_cuoi = n % 10  
    tong_chu_so += chu_so_cuoi 
    n //= 10
print(f"Tổng các chữ số của {temp} là: {tong_chu_so}")
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không

```python
n = int(input("Nhập vào số nguyên dương n: "))
temp = n
if n <= 0:
    is_power_of_three = False
else:
    while n % 3 == 0:
        n //= 3
    if n == 1:
        is_power_of_three = True
    else:
        is_power_of_three = False
if is_power_of_three:
    print(f"{temp} là số dạng 3^k.")
else:
    print(f"{temp} không phải là số dạng 3^k.")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("Nhập vào số nguyên dương A: "))
f1 = 1
f2 = 1
fib_max = 1
if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: 1")
else:
    while True:
        f_next = f1 + f2
        if f_next > A:
            fib_max = f2
            break
        f1 = f2
        f2 = f_next
        fib_max = f2
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_max}")
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input("Nhập vào số nguyên dương n: "))
temp = n
count = 0
so_nguyen_to = [2, 3, 5, 7]
while n > 0:
    chu_so = n % 10  
    if chu_so in so_nguyen_to:
        count += 1
    n //= 10 
print(f"--- Kết quả cho số {temp} ---")
print(f"Số lượng chữ số là số nguyên tố: {count}")
```
