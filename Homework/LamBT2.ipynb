Bài 3: Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
a, b, c = map(int,input().split())
if a + b > c and a + c > b and b + c > a:
    print("Tam giác hợp lệ")
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giác vuông")
    else: 
        print("Tam giác thường")
else:
    print("Không là ba cạnh của một tam giác")
p = a + b + c
s = (p/2 * (p/2 -a) * (p/2 - b) * (p/2 - c))**(1/2)
print(f"Chu vi: {p}, Diện tích: {s}")
```

Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0) 

```python
a, b = map(float,input().split())
if a != 0:
    x = -b / a
    print(f"Phương trình có một nghiệm duy nhất: x = {x}")
else:
    if b == 0:
        print("Phương trình có vô số nghiệm")
    else:
        print("Phương trình vô nghiệm")
```

```python
a, b = map(float,input().split())
match a, b: 
    case a, b if a != 0:
        x = -b / a
        print(f"Phương trình có một nghiệm duy nhất: x = {x}")
    case _:
        if b == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float,input().split())
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        print(f"Phương trình bậc nhất có nghiệm x = {-c/b}")
else:
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + delta**(1/2)) / (2*a)
        x2 = (-b - delta**(1/2)) / (2*a)
        print(f"Phương trình có hai nghiệm phân biệt: x1 = {x1}, x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép: x = {x}")
    else:
        print("Phương trình vô nghiệm")
```

Bài 8: Giải hệ phương trình tuyến tính: 
ax + by = m, cx + dy = n

```python
a, b, c, d, m, n = map(float,input().split())
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}")
else: 
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

Bài 9:  Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây.  

```python
x = float(input())
gio = x//3600
phut = (x % 3600) // 60
giay = (x % 3600) % 60
print(f"Kết quả: {gio} giờ, {phut} phút, {giay} giây")
```

Bài 10:  Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b = map(float,input("Nhập tọa độ tâm đường tròn (x, y): ").split())
r = float(input("Nhập bán kính đường tròn: "))
c, d = map(float,input("Nhập tọa độ điểm cần kiểm tra (x, y): ").split())
d = ((c - a)**2 + (d - b)**2)**(1/2)
match d, r:
    case d, r if d < r:
        print("Điểm nằm trong đường tròn")
    case d, r if d == r:
        print("Điểm nằm trên đường tròn")
    case _:
        print("Điểm nằm ngoài đường tròn")
```
